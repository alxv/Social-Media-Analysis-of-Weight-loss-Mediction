# Real-World Dosing Behaviour for GLP-1 'Skinny Jabs' (Publication Analysis)

**Research focus:** Forum discussions on Mounjaro, Wegovy, Ozempic, and Saxenda regarding cost-driven deviations from label dosing (interval stretching, stop–restart cycles, pen spreading) and co-discussion of skeletal muscle / body-composition loss.

**Run analysis:** `python scripts/run_realworld_dosing_publication_analysis.py`

**Outputs:** `data/realworld_dosing_publication/`

```python
import subprocess, sys
from pathlib import Path
import pandas as pd
from IPython.display import Image, display, Markdown

BASE = Path.cwd()
OUT = BASE / "data" / "realworld_dosing_publication"
FIG = OUT / "figures"

# Re-run analysis (optional; comment out if already generated)
subprocess.run([sys.executable, "scripts/run_realworld_dosing_publication_analysis.py"], check=True, cwd=BASE)
subprocess.run([sys.executable, "scripts/run_submission_additions.py"], check=True, cwd=BASE)

prev = pd.read_csv(OUT / "Table_brand_prevalence.csv")
sem = pd.read_csv(OUT / "Table_semaglutide_sensitivity.csv")
first = pd.read_csv(OUT / "Table_first_mentions.csv", parse_dates=["first_forum_date", "reference_approval_date"])
chi = pd.read_csv(OUT / "Table_chi_square.csv")
val_metrics = pd.read_csv(OUT / "validation" / "Table_llm_validation_metrics.csv")

display(Markdown("## Table: Behaviour prevalence by brand"))
display(prev)
display(Markdown("## Table: First forum mentions vs approval dates"))
display(first)
display(Markdown("## Semaglutide sensitivity (brand-split vs combined)"))
display(sem)
display(Markdown("## LLM validation metrics (n=200 stratified sample)"))
display(val_metrics)
display(Markdown("## Statistical comparisons (chi-square)"))
display(chi)
```
